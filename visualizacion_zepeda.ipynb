```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('../1st_datasets/movie_metadata.csv')
movies
```

```python
# Histograma
# Cuales es la duracion con mas frecuencia
fig, ax = plt.subplots(figsize=(18,6))
ax.hist(x= movies.duration, bins=30, density=False, orientation='vertical')
ax.set_xlabel('Directores')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de duracion de Peliculas')
# La frecuencia mas alta de duracion de las peliculas es de alrededor de 100 minutos
```

```python
#Pie chart data
d_pie = movies[['country','director_name']].groupby(by='country').count()
d_pie
```

```python
#Pie chart
# que pais tiene mas peliculas y que porcentaje respecto al total representa
fig,ax = plt.subplots(figsize =(12, 6))
ax.pie(d_pie.director_name, labels = d_pie.index,autopct='%.1f%%', )
ax.set_title('Piechart de distribucion de peliculas por pais')
#USA es el pais con mas peliculas, acaparando el 75.6% del dataset
```

```python
# Barplot
bplot_data = movies.director_name.value_counts()
f_bplot_data = bplot_data[bplot_data > 10]
```

```python
# Cuales directores tienen mas de diez peliculas
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(f_bplot_data.index, f_bplot_data)
ax.set_title('Numero de peliculas por director')
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Director')
ax.set_xticklabels(f_bplot_data.index, rotation='vertical')
plt.show()

#Steven Splielberg ess el director con mas peliculas
```

```python
# Multiples figuras 
# Cual es el puntaje que mas se repite en los generos de Drama, Comedia y Comedia-Drama en peliculas de USA
m1 = movies[(movies['country'] == 'USA') & (movies['genres'].isin(['Drama', 'Comedy', 'Comedy|Drama']))]
sns.displot(data=m1, x="imdb_score", hue="genres", col="genres")

# El score que mas serepite para comedia entre 6 y 7, para Comedia-Drama y Drame es entre 7 y 7.5
```

```python
# Existe alguna relacion con el presupuesto de la pelicula con su calificaion dependiendo del genero

f, axs = plt.subplots(1, 2, figsize=(14, 8), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=m1, x="imdb_score", y="budget", hue="genres", ax=axs[0])
sns.histplot(data=m1, x="genres", hue="genres", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()

# En general, las peliculas de drama tienen mejores calificaciones, sin embargo no existe una clara relacion de este desempeno con el presupuesto invertido en ellas. 
```
